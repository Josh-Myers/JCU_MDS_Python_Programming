```
# example urls
# occupation only: 'https://jobsearch.gov.au/job/search/external?Occupation=6113'
# occupation and location: https://jobsearch.gov.au/job/search/external?Occupation=2613&Location=19AACQ 
url_occ = 'https://jobsearch.gov.au/job/search/external?Occupation='
url_loc = '&Location='

urls = []
for job in job_ids.values():
  for location in locations.values():
    urls.append(url_occ + job + url_loc + location)
    
```

```
urls = random.sample(urls, 200)
```

```
%%time
# use wait instead of sleep to make more efficient https://stackoverflow.com/questions/25922652/selenium-webdriver-for-python-trouble-with-loops
# extract element from page
counts = []
for url in urls:
  driver = webdriver.Chrome('chromedriver',options=options)
  time.sleep(1)
  driver.get(url)
  time.sleep(2)
  count = driver.find_elements_by_xpath('.//div[@class="job-search-filter-tags search-filter-tags"]')
  counts.append(count[0].text)
  
# extract job and position counts
out = []
for count in counts:
  nums = [int(c) for c in count.split() if c.isdigit()]
  out.append(nums)
    
jobs = [i[0] for i in out]
positions = [i[1] for i in out]  

# job and location labels
labels = []
for job in job_ids.values():
  for location in locations.keys():
    id = job + "_" + location
    labels.append(id)
    
job_labs = []
location_labs = []
for l in labels:
  labs = l.split("_")
  job_labs.append(labs[0])
  location_labs.append(labs[1])

# save results to data frame and export
jobsearch_counts = pd.DataFrame(list(zip(job_labs, location_labs, jobs, positions)), columns = ['anzsco', 'location', 'number_jobs', 'number_positions'])
jobsearch_counts.to_csv("jobsearch_counts.csv", index=False)
files.download("jobsearch_counts.csv")
```
